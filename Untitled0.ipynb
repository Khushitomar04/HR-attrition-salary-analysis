```python
import pandas as pd
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df['SalaryBand'] = pd.cut(df['MonthlyIncome'],
                           bins=[0, 3000, 6000, 10000, 20000],
                           labels=['Low', 'Medium', 'High', 'Very High'])

df[['MonthlyIncome', 'SalaryBand']].head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Chart 1: Attrition by Salary Band
plt.figure(figsize=(8,5))
sns.countplot(x='SalaryBand', hue='Attrition', data=df, order=['Low','Medium','High','Very High'])
plt.title('Attrition by Salary Band')
plt.xlabel('Salary Band')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# calculate percentange attrition
attrition_pct = df.groupby('SalaryBand')['Attrition'].apply(lambda x: (x=='Yes').mean()*100)

plt.figure(figsize=(8,5))
attrition_pct.plot(kind='bar', color='indianred')
plt.title('Attrition Rate (%) by Salary Band')
plt.xlabel('Salary Band')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)
plt.show()
```

```python
# chart-2 Attrition by department
grouped_dept = df.groupby(['Department', 'Attrition']).size().unstack()

grouped_dept.plot(kind='bar', figsize=(8,5), color=['skyblue', 'salmon'])
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=15)
plt.legend(title='Attrition')
plt.show()
```

```python
#chart-3 Attrition by overtime
grouped_ot = df.groupby(['OverTime', 'Attrition']).size().unstack()


grouped_ot.plot(kind='bar', figsize=(8,5), color=['skyblue', 'salmon'])
plt.title('Attrition by OverTime')
plt.xlabel('OverTime')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.legend(title='Attrition')
plt.show()
```

```python
#chart-4 Average Monthly Income
avg_salary = df.groupby('Attrition')['MonthlyIncome'].mean()

plt.figure(figsize=(6,5))
avg_salary.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Average Monthly Income: Stayed vs Left')
plt.xlabel('Attrition')
plt.ylabel('Average Monthly Income')
plt.xticks(rotation=0)
plt.show()
```

## 📊 Key Insights - HR Attrition & Salary Analysis

1. **Salary Impact**: Employees in the Low salary band have the highest attrition rate (~29%), while the Very High salary band has only ~9% attrition. This shows a clear inverse relationship between salary and attrition.

2. **Department Risk**: The Sales department has the highest attrition rate compared to Research & Development and Human Resources — likely due to high-pressure, target-driven roles.

3. **OverTime Effect**: Employees who work overtime have an attrition rate (approximately 31%) that is three times higher than employees who don't work overtime (approximately 10%).

4. **Average Salary Gap**: Employees who left the company had a significantly lower average monthly income (approximately Rs. 4800) compared to employees who stayed (approximately Rs. 6800) — a gap of around Rs. 2000.

## Recommendations

- Design a targeted retention plan for employees in the Low salary band and the Sales department.
- Review the overtime policy — consider workload balancing or additional compensation for employees who consistently work overtime.
- Conduct salary benchmarking to ensure a competitive pay structure, especially for entry-level roles.
